## コードにおける辞書の例

下のセルでは、上の例のデータを格納するために、ハッシュテーブル（`unordered_map`）を作成しています。C++で `unordered_map` を作成するには、`<unordered_map>` ヘッダーを含める必要があり、`unordered_map` を宣言する構文は以下のようになります:
```cpp
unordered_map <key_type, value_type> variable_name;
```
以下のコードでは、`.find()`メソッドを用いて、`key` が `unordered_map` に存在するかどうかをチェックしています。key がマップに存在しない場合は、`.find()` は`unordered_map::end()`型を返します。そうでなければ、`.find()`は [C++ iterator](http://www.cplusplus.com/reference/iterator/) を返します。これはイテレート可能なキーと値のペアの先頭を指すポインタです。

このコースではまだイテレータを扱っておらず、このプロジェクトでも必要ありませんが、イテレータは範囲内で前方または後方に"反復）"できるポインタによく似ています。

````c++
#include <iostream>
#include <vector>
#include <unordered_map>
#include <string>
using std::vector;
using std::cout;
using std::unordered_map;
using std::string;


// ハッシュテーブルで使用する文字列を作成。
string key = "word";
string def_1 = "a unit of language, consisting of one or more spoken sounds or their written representation, that functions as a principal carrier of meaning";
string def_2 = "speech or talk: to express one's emotion in words";
string def_3 = "a short talk or conversation: 'Marston, I'd like a word with you.'";
string def_4 = "an expression or utterance: a word of warning";
unordered_map <string, vector<string>> my_dictionary;

// キーがハッシュテーブルにあるかどうかをチェック。
if (my_dictionary.find(key) == my_dictionary.end()) {
    cout << "The key 'word' is not in the dictionary." << "\n";
    cout << "Inserting a key-value pair into the dictionary." << "\n\n";
    // キーの値を設定。
    my_dictionary[key] = vector<string> {def_1, def_2, def_3, def_4};
}

// これでキーがハッシュテーブルに入るはずです。
// キーに対応する値にアクセスするには、角括弧 [] を使います。
// ここでは、my_dictionary[key] という値は文字列のベクトルです。
// ベクターを反復処理して、文字列を印刷します。
cout << key << ": \n";
auto definitions = my_dictionary[key];
for (string definition : definitions) {
    cout << definition << "\n";
}
````
